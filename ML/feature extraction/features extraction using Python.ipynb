# Autocorrelation descriptors positive

```python
from Bio import SeqIO
import numpy as np
import csv

def calculate_autocorrelation(sequence, lag):
    n = len(sequence)
    autocorr = np.zeros(lag)
    
    for k in range(1, lag+1):
        sum_ac = 0
        for i in range(n - k):
            sum_ac += (sequence[i] == sequence[i+k])
        autocorr[k-1] = sum_ac / (n - k)
    
    return autocorr

filename = "C:/Users/ADMIN/ML major project/positive.fasta"
text_output_file = "C:/Users/ADMIN/ML major project/positive_autocorr_output.txt"
csv_output_file = "C:/Users/ADMIN/ML major project/positive_autocorr_output.csv"
lag = 5  # Maximum lag for autocorrelation calculation

# Open the text output file in write mode
with open(text_output_file, "w") as text_file:
    # Open the CSV output file in write mode
    with open(csv_output_file, "w", newline='') as csv_file:
        writer = csv.writer(csv_file)
        
        # Parse the fasta file
        for record in SeqIO.parse(filename, "fasta"):
            # Calculate autocorrelation
            autocorr = calculate_autocorrelation(str(record.seq), lag)
            
            # Save autocorrelation values as text
            text_line = record.id + "\t" + "\t".join(map(str, autocorr)) + "\n"
            text_file.write(text_line)
            
            # Save autocorrelation values as CSV
            csv_row = [record.id] + list(autocorr)
            writer.writerow(csv_row)
```

```python
from Bio import SeqIO
import numpy as np
import csv

def calculate_normalized_mb_autocorrelation(sequence, lag):
    n = len(sequence)
    normalized_mb_autocorr = np.zeros(lag)
    
    mean = np.mean(sequence)
    std = np.std(sequence)
    
    for k in range(1, lag+1):
        sum_ac = 0
        for i in range(n - k):
            sum_ac += (sequence[i] - mean) * (sequence[i+k] - mean)
        normalized_mb_autocorr[k-1] = sum_ac / ((n - k) * std**2)
    
    return normalized_mb_autocorr

filename = "C:/Users/ADMIN/ML major project/positive.fasta"
text_output_file = "C:/Users/ADMIN/ML major project/positive_normalized_mb_autocorr_output.txt"
csv_output_file = "C:/Users/ADMIN/ML major project/positive_normalized_mb_autocorr_output.csv"
lag = 5  # Maximum lag for autocorrelation calculation

# Open the text output file in write mode
with open(text_output_file, "w") as text_file:
    # Open the CSV output file in write mode
    with open(csv_output_file, "w", newline='') as csv_file:
        writer = csv.writer(csv_file)
        
        # Parse the fasta file
        for record in SeqIO.parse(filename, "fasta"):
            # Convert sequence to numerical values
            sequence = np.array([ord(aa) for aa in str(record.seq)])
            
            # Calculate normalized Moreau-Broto autocorrelation
            normalized_mb_autocorr = calculate_normalized_mb_autocorrelation(sequence, lag)
            
            # Save normalized Moreau-Broto autocorrelation values as text
            text_line = record.id + "\t" + "\t".join(map(str, normalized_mb_autocorr)) + "\n"
            text_file.write(text_line)
            
            # Save normalized Moreau-Broto autocorrelation values as CSV
            csv_row = [record.id] + list(normalized_mb_autocorr)
            writer.writerow(csv_row)
```

# Moran autocorrelation descriptors positive

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
from Bio import SeqIO

def get_moran_autocorrelation(sequence):
    numerical_sequence = np.array([ord(aa) for aa in sequence], dtype=float)
    autocorrelation = pdist(numerical_sequence.reshape(-1, 1), metric='correlation')
    distance_matrix = squareform(autocorrelation)

    descriptors = []
    for lag in range(1, len(sequence)):
        descriptor = np.mean(distance_matrix[np.diag_indices(len(sequence)-lag)])
        descriptors.append(descriptor)

    return descriptors

filename = "C:/Users/ADMIN/ML major project/positive.fasta"  # Replace with the path to your positive.fasta file
moran_descriptors = []

# Parse the fasta file
for record in SeqIO.parse(filename, "fasta"):
    sequence = record.seq
    descriptors = get_moran_autocorrelation(sequence)
    moran_descriptors.append([record.id] + descriptors)

# Save the descriptors to a text file
output_file = "C:/Users/ADMIN/ML major project/positive_moran_output.txt"  # Replace with the desired output file path
with open(output_file, "w") as file:
    for descriptor in moran_descriptors:
        file.write(f"{','.join(map(str, descriptor))}\n")

# Save the descriptors to a CSV file
csv_file = "C:/Users/ADMIN/ML major project/positive_moran_output.csv"  # Replace with the desired output CSV file path
import csv
with open(csv_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Sequence ID"] + [f"Lag_{i}" for i in range(1, len(moran_descriptors[0]))])
    writer.writerows(moran_descriptors)
```

# Geary autocorrelation descriptors

```python
from Bio import SeqIO
import numpy as np
from scipy.spatial.distance import pdist, squareform

def encode_sequence(sequence):
    encoding_dict = {'A': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5, 'G': 6, 'H': 7, 'I': 8, 'K': 9, 'L': 10,
                     'M': 11, 'N': 12, 'P': 13, 'Q': 14, 'R': 15, 'S': 16, 'T': 17, 'V': 18, 'W': 19, 'Y': 20}
    encoded_sequence = [encoding_dict[aa] for aa in sequence]
    return encoded_sequence

def get_geary_autocorrelation(sequence):
    encoded_sequence = encode_sequence(sequence)
    sequence_array = np.array(encoded_sequence)
    autocorrelation = pdist(sequence_array.reshape(-1, 1), metric='seuclidean')
    distance_matrix = squareform(autocorrelation)
    n = len(sequence_array)
    geary_descriptors = []
    for lag in range(1, n):
        numerator = ((distance_matrix[:-lag, lag:] ** 2).sum()) / (2 * (n - lag))
        denominator = 2 * ((distance_matrix[:-lag, lag:] ** 2).sum()) / ((n - lag) * (n - lag + 2))
        geary_descriptors.append(numerator / denominator)
    return geary_descriptors

# Read protein sequences from a FASTA file
filename = "C:/Users/ADMIN/ML major project/positive.fasta"
sequences = [str(record.seq) for record in SeqIO.parse(filename, "fasta")]

# Calculate Geary autocorrelation descriptors for each sequence
geary_descriptors_list = []
for sequence in sequences:
    geary_descriptors = get_geary_autocorrelation(sequence)
    geary_descriptors_list.append(geary_descriptors)

# Print the descriptors for each sequence
for i, geary_descriptors in enumerate(geary_descriptors_list):
    print(f"Sequence {i+1} Geary Autocorrelation Descriptors:")
    print(geary_descriptors)
```
