# counting the clusters

```python
def count_sequences_in_clusters(clstr_file):
    with open(clstr_file, 'r') as f:
        lines = f.readlines()

    sequence_counts = []
    for line in lines:
        if line.startswith('>Cluster'):
            if sequence_counts:
                print(f'Cluster {len(sequence_counts)} has {sequence_counts[-1]} sequences')
            sequence_counts.append(0)
        else:
            sequence_counts[-1] += 1

    # Print count for the last cluster
    print(f'Cluster {len(sequence_counts)} has {sequence_counts[-1]} sequences')

    return sequence_counts

# Usage
sequence_counts = count_sequences_in_clusters('negative_seq_cluster_0.99.clstr')  # replace with your .clstr file
```
